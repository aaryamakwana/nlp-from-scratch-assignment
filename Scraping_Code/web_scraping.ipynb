### Encyclopedia Brittanica page

```python


# URL of the webpage to scrape
url = "https://www.britannica.com/place/Pittsburgh"

# Send an HTTP GET request to the webpage
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all sections with headers and their content
sections = soup.find_all(['h2', 'h3', 'h4'])  # Adjust tags based on actual header tags used

# List to hold dictionaries with section titles and content
content_list = []

for section in sections:
    # Get the title of the section
    title = section.get_text()
    
    # Get the content following the section header until the next header or end of document
    content = []
    for sibling in section.find_next_siblings():
        if sibling.name in ['h2', 'h3', 'h4']:  # Stop if another header is encountered
            break
        content.append(sibling.get_text())
    
    # Join content list into a single string
    content_text = ' '.join(content)
    
    # Append dictionary to list
    content_list.append({"title": title, "content": content_text})

# Save the list of dictionaries to a pickle file
with open('pittsburgh_content.pkl', 'wb') as file:
    pickle.dump(content_list[1:], file)

print("Content successfully scraped and saved to pittsburgh_content.pkl")
```

### City of Pittsburgh Tax Regulations: See the links under the "Regulations" column of the table

```python
# i downloaded all the pdfs from the website and saved them in the folder
pdfs_path = "pdfs_extracted/"

title_to_file_path  = {"Amusement Tax Regulations":"pdfs_extracted/amusement_tax_regulations.pdf",
                       "Payroll Expense Tax Regulations":"pdfs_extracted/payroll_tax_regulations.pdf",
                       "Institution & Service Privilege Regulations":"pdfs_extracted/THE_INSTITUTION_AND_SERVICE_PRIVILEGE_TAX_REGULATIONS.pdf",
                       "Local Services Tax-Employers and Self-Employed Regulations":"pdfs_extracted/local_services_tax.pdf",
                       "Parking Tax Regulations":"pdfs_extracted/parking_tax_regulations.pdf",
                       "Non Resident Sports Facility Usage Fee Regulations":"pdfs_extracted/UF_Regulations.pdf",
                       "Operating Budget": "pdfs_extracted/23255_2024_Operating_Budget.pdf"}


# Extract text from the PDFs
# create a list of dictionaries with structure as [{"title": , "content": },{"title": , "content": }{"title": , "content": }]
pdf_content_list = []

for title, file_path in title_to_file_path.items():
    # Read the PDF file
    pdf = PdfReader(file_path)
    
    # Extract text from each page
    content = []
    for page in pdf.pages:
        content.append(page.extract_text())
    
    # Join content list into a single string
    content_text = ' '.join(content)
    
    # Append dictionary to list
    pdf_content_list.append({"title": title, "content": content_text})

# Save the list of dictionaries to a pickle file
with open('files_scraped/pittsburgh_tax_regulations.pkl', 'wb') as file:
    pickle.dump(pdf_content_list, file)
```

## About CMU & CMU History

```python
# List of URLs to scrape
urls = [
    "https://www.cmu.edu/about/",
    "http://www.cmu.edu/leadership/",
    "https://www.cmu.edu/about/mission.html",
    "https://www.cmu.edu/about/history.html",
    "https://www.cmu.edu/about/traditions.html",
    "http://www.cmu.edu/diversity",
    "https://www.cmu.edu/about/pittsburgh.html",
    "https://www.cmu.edu/about/rankings.html",
    "https://www.cmu.edu/about/awards.html"
]

# List to hold the scraped content
content_list = []

for url in urls:
    # Send a GET request to each URL
    response = requests.get(url)
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the title of the page
    title = soup.title.string if soup.title else 'No Title'
    
    # Extract the main content (adjust selector based on actual structure)
    main_content = soup.find('main') or soup.find('body')
    
    if main_content:
        content_text = main_content.get_text(separator=' ', strip=True)
        content_list.append({"title": title, "content": content_text})

# Save the list of dictionaries to a JSON file
with open('cmu_about_content.json', 'w', encoding='utf-8') as file:
    json.dump(content_list, file, ensure_ascii=False, indent=4)

print("Content successfully scraped and saved to cmu_about_content.json")
```

## [Visit Pittsburgh website](https://www.visitpittsburgh.com/)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import json

# Initialize the Selenium WebDriver
driver = webdriver.Chrome()  # Ensure you have the ChromeDriver installed and added to PATH

# List of URLs to scrape
urls = [
    "https://www.visitpittsburgh.com/",
    "https://www.visitpittsburgh.com/things-to-do/family-fun/?hitsPerPage=96",
    "https://www.visitpittsburgh.com/things-to-do/arts-culture/?hitsPerPage=96",
    "https://www.visitpittsburgh.com/things-to-do/arts-culture/?hitsPerPage=96&page=2",
    "https://www.visitpittsburgh.com/restaurants-culinary/?hitsPerPage=96",
    "https://www.visitpittsburgh.com/restaurants-culinary/?hitsPerPage=96&page=2",
    "https://www.visitpittsburgh.com/events-festivals/annual-events/",
    "https://www.visitpittsburgh.com/things-to-do/pittsburgh-sports-teams/"
]

# List to hold the scraped content
content_list = []

def scrape_page(url):
    driver.get(url)
    
    # Wait for the page content to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, 'body'))
    )
    
    # Use BeautifulSoup to parse the page content
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Extract information cards if present
    cards = soup.find_all('div', class_='card__inner')
    
    for card in cards:
        heading = card.find('a', class_='card_heading').get_text(strip=True) if card.find('a', class_='card_heading') else 'No Heading'
        summary = card.find('div', class_='card_summary').get_text(strip=True) if card.find('div', class_='card_summary') else 'No Summary'
        address = card.find('div', class_='card_address').get_text(strip=True) if card.find('div', class_='card_address') else 'No Address'
        phone = card.find('div', class_='card_phone').get_text(strip=True) if card.find('div', class_='card_phone') else 'No Phone'
        
        content_list.append({
            "heading": heading,
            "summary": summary,
            "address": address,
            "phone": phone
        })
    
    # For pages without specific cards, just get the main content
    if not cards:
        main_content = soup.find('main') or soup.find('body')
        if main_content:
            title = soup.title.string if soup.title else 'No Title'
            content_text = main_content.get_text(separator=' ', strip=True)
            content_list.append({"title": title, "content": content_text})

for url in urls:
    scrape_page(url)

# Save the list of dictionaries to a JSON file
with open('visit_pittsburgh_content.json', 'w', encoding='utf-8') as file:
    json.dump(content_list, file, ensure_ascii=False, indent=4)

print("Content successfully scraped and saved to visit_pittsburgh_content.json")

# Close the browser
driver.quit()
```
